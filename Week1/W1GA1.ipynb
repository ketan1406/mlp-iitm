## Q6

Which of the following features have missing(NaN) values present in the dataset?
(Note: compute after removing "?")

```python
# Check for NaN values in each feature
missing_values = df.isna().sum()
features_with_missing = missing_values[missing_values > 0]
print(missing_values)
# Display features with missing values
if not features_with_missing.empty:
    print("Features with missing values (NaN):")
    for feature, count in features_with_missing.items():
        print(f"- {feature}: {count} missing values")
else:
    print("No features have missing values!")
```

## Q7

Which of the following feature has most missing(NaN) values present in the dataset?
 (Note: compute after removing "?")

```python
missing_values1 = df.isna().sum()

# Find the feature(s) with the maximum NaN count
max_missing = missing_values1.max()
features_with_max = missing_values[missing_values == max_missing].index.tolist()

# Display the result
if max_missing == 0:
    print("No features have missing values!")
else:
    if len(features_with_max) == 1:
        print(f"Feature with the most missing values: '{features_with_max[0]}' ({max_missing} NaN values)")
    else:
        print(f"Features tied with the most missing values ({max_missing} NaN values each):")
        print(features_with_max)
```

## Q6

Determine the number of samples (rows) in the dataset that belong to the six most recent years, based on the provided data.

```python
# Convert the year column to numeric (replace 'Year' with your actual column name)
year_column = 'Year'  # Update this to match your dataset
df[year_column] = pd.to_numeric(df[year_column], errors='coerce')

# Extract unique non-null years and sort them in descending order
unique_years6 = df[year_column].dropna().unique()
sorted_years = sorted(unique_years6, reverse=True)

# Select the six most recent years (or fewer if <6 exist)
top_years = sorted_years[:6]

# Filter rows where the year is in the top six
filtered_df6 = df[df[year_column].isin(top_years)]

# Count the remaining samples
sample_count = filtered_df6.shape[0]

print(f"Number of samples in the six most recent years: {sample_count}")
```

## Q10

Calculate the number of samples(rows) that correspond to the month of August across all the years.?

```python
# Convert date column to datetime format (replace 'date' with actual column name)
date_column = 'Date'  # Update with your column name
df[date_column] = pd.to_datetime(df[date_column], errors='coerce')

# Extract month from date
df['month'] = df[date_column].dt.month

# Filter August entries (month == 8)
august_samples = df[df['month'] == 8]

# Count August samples
august_count = august_samples.shape[0]

print(f"Number of samples from August across all years: {august_count}")
```

Q11

According to the data which Locality have highest average Sale Price?

```python
# Clean data: Handle missing values and numeric conversion
df['Sale Price'] = pd.to_numeric(df['Sale Price'], errors='coerce')
df.dropna(subset=['Sale Price'], inplace=True)  # Remove rows with NaN sale prices

# Calculate average sale price per locality
avg_prices = df.groupby('Locality')['Sale Price'].mean()

# Find locality with highest average
max_locality = avg_prices.idxmax()
max_price = avg_prices.max()

print(f"Locality with highest average sale price: {max_locality}")
print(f"Average price: ${max_price:,.2f}")
```
